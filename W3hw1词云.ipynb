```python
import matplotlib.pyplot as plt
import random

words = ["Python", "NumPy", "Pandas", "matplotlib", "Scikit-learn"]
counts = [random.randint(1, 100) for i in range(len(words))] 

fig, ax = plt.subplots()
ax.scatter([i for i in range(len(words))], counts, marker='o')

for i, word in enumerate(words):
    ax.text(i, counts[i], word, fontsize=counts[i], ha='center')
    
ax.set_title("Word Cloud")
ax.set_xlabel("Word Index")
ax.set_ylabel("Count")
ax.grid()

plt.show()
```
