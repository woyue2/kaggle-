np.random.randn 的 randn 是 random <b>normal</b> distribution
```python
Two-by-four array of samples from N(3, 6.25):
    2.5 ** 2 = 6.25 所以是 3 + 2.5 ellipis
    mean is 3, standard deviation标准差2.5
    方差variance。6.25
    正态分布（normal distribution）
    
>>> 3 + 2.5 * np.random.randn(2, 4)
array([[-4.49401501,  4.00950034, -1.81814867,  7.29718677],   # random
       [ 0.39924804,  4.68456316,  4.99394529,  4.84057254]])  # random
```

reindex会填充 NaN如果 index在之前不存在的话

````python
df6 = pd.DataFrame(np.random.randn(5,3),
                  index = list('acefh'),
                  columns= ['one', 'two', 'three']
                  )
df6
````

````python
df7 = df6.reindex(list('abcdefgh'))
df7
````

````python
df6['one'].isnull()
````

````python
res = df7['one'].isnull()
res
type(res)
````

````python
df7['one'].sum() # 
````

````python
list = pd.Series(np.array([1, None])) 
# NumPy数组的数据类型是object，因为它可以容纳不同类型的对象。因此，Series的数据类型也是object。
# 在这种情况下，Series的value实际上是一个Python对象，即None。它是Python中的一个内置常量，属于NoneType类型。
list
````

````python
list.sum()
````

````python
res2 = pd.Series([1, None]) # 默认情况下，Pandas将None解释为缺失值（NaN），它是一个浮点数类型。
res2
````

````python
res2.sum()
````
