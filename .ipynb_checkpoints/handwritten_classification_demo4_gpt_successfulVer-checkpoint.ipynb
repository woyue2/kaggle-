```python
import numpy as np
import random
learning_rate = 1e-4
```

```python
def readOneLine(content : str):
    content = content[0:-1]
    lst = content.split(',')
    for i in range(len(lst)):
        if lst[i] == '?':
            return []
        else:
            lst[i] = int(lst[i])
        id, f1, f2, f3, f4, f5, f6, f7, f8, f9, state = \
        lst[0], lst[1], lst[2], lst[3], lst[4], lst[5], lst[6], lst[7], lst[8], lst[9], lst[10]
    return [id, f1, f2, f3, f4, f5, f6, f7, f8, f9, state]

def getX(data : list):
    x0 = 1
    x1 = data[1]
    x2 = data[2]
    x3 = data[3]
    x4 = data[4]
    x5 = data[5]
    x6 = data[6]
    x7 = data[7]
    x8 = data[8]
    x9 = data[9]   
    return [x0, x1, x2, x3, x4, x5, x6, x7, x8, x9]

def sigmoid(z):
    print(f'sigmoid')
    return 1 / (1 + np.exp(-z))

def getLoss(sig_y_pred, y):
    res = -np.mean(y * np.log(sig_y_pred) + (1 - y) * np.log(1 - sig_y_pred))
    print(f'getLoss {res}')
    return res

def getPartialDerivative(y_train, sig_y_pred, y_pred, X_train):
    print('getPartialDerviative')
    LtoY = (-1* (y_pred)) * (1/ sig_y_pred) + (1- y_pred) * (1/ (1 - sig_y_pred))  # (599,)
    YtoZ = (-1) *np.exp(-sig_y_pred) / (1 - np.exp(-sig_y_pred))**2 # (599,)
    ZtoW = X_train # (599,10)
    res = (LtoY * YtoZ).T.dot(X_train)
    assert (len(res) == 10)
    print(f'{res}')
    return res

def updateWeight(lst_w, y_train, sig_y_pred, y_pred, X_train, learing_rate = 1e-4):
    print(f'lr is {learning_rate}')
    lst_w -= learning_rate * getPartialDerivative(
        y_train, 
        sig_y_pred, 
        y_pred, 
        X_train,
    ) # 向量 -= 梯度下降的向量    形式：矩阵
    print(f'update w :\n{lst_w}' )
    return lst_w
```

```python
# read data
file = open("data.csv", "r")
content = file.readlines()
data = []
for i in range(len(content)):
    data.append(readOneLine(content[i]))
data_fil = []
for i in range(len(data)):
    if data[i]:
        data_fil.append(data[i])
data = data_fil

# split X_train, y_train, X_test, y_test
X = []
for i in range(len(data)):
    lst = getX(data[i])
    X.append(lst)
Y = []
for i in range(len(data)):
    if (data[i][-1]) == 2: 
        Y.append(0) # 良性
    else:
        Y.append(1) # 阴性
        
X_train, X_test = np.array(X[0:599]), np.array(X[600:])
y_train, y_test = np.array(Y[0:599]), np.array(Y[600:])
```

```python
class LogisticRegression:
    def __init__(self, learning_rate=0.01, num_iterations=1000, stop_threshold=0.01):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.stop_threshold = stop_threshold
        self.weights = None
        self.bias = None

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
    
    def compute_loss(self, y, y_pred):
        epsilon = 1e-5  # 用于防止log(0)的情况
        loss = -np.mean(y * np.log(y_pred + epsilon) + (1 - y) * np.log(1 - y_pred + epsilon))
        return loss
    
    def fit(self, X, y):
        num_samples, num_features = X.shape

        # 初始化权重和偏置
        self.weights = np.zeros(num_features)
        self.bias = 0

        for iteration in range(self.num_iterations):
            linear_model = np.dot(X, self.weights) + self.bias
            y_pred = self.sigmoid(linear_model)

            # 计算损失函数
            loss = self.compute_loss(y, y_pred)

            # 计算梯度
            dw = (1 / num_samples) * np.dot(X.T, (y_pred - y))
            db = (1 / num_samples) * np.sum(y_pred - y)

            # 更新权重和偏置
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

            # 打印损失函数
            print(f"Iteration {iteration+1}, Loss: {loss}")

            # 判断是否满足停止准则
            if loss < self.stop_threshold:
                print("Reached stop threshold. Stopping training.")
                break
                
            print(f'{self.weights} {self.bias}\n')
            
    def predict(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        print(f"{linear_model}")
        y_pred = self.sigmoid(linear_model)
        y_pred_class = np.where(y_pred > 0.5, 1, 0)
        return y_pred_class
```

```python
X = X_train
y = y_train
model = LogisticRegression(learning_rate=0.1, num_iterations=1000, stop_threshold=0.01)
model.fit(X, y)
```
