```python
import sympy as sp
from dataclasses import dataclass
```

```python
from sympy import Symbol, Derivative, Function
from sympy.abc import x, w, b
```

```python
@dataclass
class Point():
    x : int
    y : int
    
    def __repr__(self):
        return f'coordinate is ({self.x},{self.y})'
@dataclass
class Regression():
    weig : int
    bais : int
    def __repr__(self):
         return ( f'formula is {self.weig} * x + {self.bais}'  )        

```

```python
y_pred = sp.symbols('y_pred', cls = Function)
y_pred = w * x + b
y_pred 
```

```python
p = Point(1.5, 0.8)
f = Regression(0.8, 0.2)
```

```python
# 一条线 一个点 start
```

```python
Loss = sp.symbols('Loss', cls = Function)
# prediction = y_pred
# target = p.y
Loss = 1/2 * (y_pred - p.y)**2
Loss
```

```python
Partial1 = Derivative(Loss, 'w', evaluate = True)
Partial2 = Derivative(Loss, 'b', evaluate = True)
# Partial2.subs({x:p.x, w:f.weig, b:f.bais}) # 0.6
```

```python
# learning rate
lr = 0.1  
```

```python
def update():
    f.weig, f.bais = f.weig - lr * Partial1.subs({x:p.x, w:f.weig, b:f.bais}), \
                f.bais- lr* Partial2.subs({x:p.x, w:f.weig, b:f.bais})
```

```python
target = p.y
counter = 0
y_p = y_pred.subs({x:p.x, w:f.weig, b:f.bais})
while (y_p - target > 1e-10):
    update()
    y_p = y_pred.subs({x:p.x, w:f.weig, b:f.bais})
    counter += 1
```

```python
p1 = Point(1.5, 0.8)
p2 = Point(1.2, 0.6)
plst = [p1, p2]
f = Regression(0.8, 0.2)
lr = 0.01  # learning rate

y_pred = w * x + b

def updateLoss(target : int):
    global Partial1, Partial2
    Loss = 1/2 * (y_pred - target)**2
    Partial1 = Derivative(Loss, 'w', evaluate = True)
    Partial2 = Derivative(Loss, 'b', evaluate = True)
    return Loss.subs({x:p.x, w:f.weig, b:f.bais})
```

```python
def update(f):
    global Partial1, Partial2
    f.weig, f.bais = f.weig - lr * Partial1.subs({x:p.x, w:f.weig, b:f.bais}), \
                f.bais- lr* Partial2.subs({x:p.x, w:f.weig, b:f.bais})
    f = Regression(f.weig, f.bais)
    return f
```

```python
def sum_Loss(plst):
    sum = 0
    for p in plst:
        sum += updateLoss(p.y)
    return sum
```

```python
counter = 0
loss_sum = sum_Loss(plst)
while (loss_sum > 1e-10):
    f = update(f)
    loss_sum = sum_Loss(plst)
    print(f.weig)
    counter += 1
# f.weig =  0.430769230769232循环出不去
# 最佳点在0.5
```
